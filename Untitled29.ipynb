```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# i) Create the Text using cv2.putText
img1 = np.zeros((300, 600), dtype='uint8')
font = cv2.FONT_ITALIC
img2 = cv2.putText(img1, "Kavinilavan ", (5, 100), font, 3, (255, 0, 0), 5, cv2.LINE_AA)

# Display using matplotlib instead of cv2.imshow()
plt.imshow(img2, cmap='gray')
plt.title("Original Image")
plt.axis('off')  # Hide axes
plt.show()

# ii) Create the structuring element
kernel1 = cv2.getStructuringElement(cv2.MORPH_RECT, (21, 11))
kernel2 = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 9))

# iii) Use Opening operation
img4 = cv2.morphologyEx(img1, cv2.MORPH_OPEN, kernel2)
plt.imshow(img4, cmap='gray')
plt.title("Opening")
plt.axis('off')  # Hide axes
plt.show()

# iv) Use Closing Operation
img3 = cv2.morphologyEx(img1, cv2.MORPH_CLOSE, kernel1)
plt.imshow(img3, cmap='gray')
plt.title("Closing")
plt.axis('off')  # Hide axes
plt.show()
```
